```python
### Stage-wise Notch receptor dotplots ###

# This notebook displays the saved LIANA Notch receptor dotplots across developmental stages.
# The plots show interactions involving NOTCH1, NOTCH2, and NOTCH3 receptors across Phenograph clusters.
```

```python
# Load Libraries
from pathlib import Path
from IPython.display import display, Markdown, Image
```

```python
# Define Paths
parent_output_dir = Path("liana_stagewise_results_phenograph")

stages_to_show = ["st13", "st16", "st18", "st20", "st22", "st24", "st27"]
```

```python
### 
```

```python
for stage in stages_to_show:
    display(Markdown(f"## {stage} — Notch receptor dotplot"))

    plot_path = parent_output_dir / stage / f"{stage}_notch_receptor_dotplot_liana_phenograph.png"

    if plot_path.exists():
        display(Image(filename=str(plot_path)))
    else:
        print(f"Missing plot: {plot_path}")
```
